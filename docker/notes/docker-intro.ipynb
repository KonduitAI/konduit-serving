### Networking Concepts/Terms

* Network: 1 or more intra connected computers able to share information

* IP Address: Unique number representing a computer on a network 
DNS (Domain Name Resolution): An aliasing system enabling the use of a more human friendly alias of an IP address.

* Host Name: A human readable alias (like google.com) representing a computer’s IP address on the internet

* Loopback address: An alias (localhost for host name, 127.0.0.1 for local ip address alias)


* Request: Represents a request from a client to a server to achieve some result (like getting a webpage)

* Response: A result computed by a server as a result of a request (the actual web page created by the server)

* Host: An individual computer on a network

* Socket: A program that allows remote connections on a host

* Port: A number (1-65535) on a host that allows remote connections by a client listening on a socket

* Client/Server Architecture: A way of writing software separating logic in to clients that request responses from servers


* Protocol: A set of conventions representing how a client and server should communicate
* URL: A specialized URI specifically for HTTP 
* URI: Uniform Resource Indicator, a unique combination of host name and port with a specified protocol (http,ftp,..)

### Docker/Containers Terms

* Virtual Machines: A way of running a whole operating system in a self-contained environment emulating hardware using software

* Container: Originally a linux concept that’s since been overloaded to include an abstraction/spec for running an individual “host” generally a linux OS (but can be windows as well), typically thought of as lighter weight virtual machines

* Docker: An implementation of the Open Container Interface (original implementation now called moby) that facilitates packaging and execution of containers

* Open Container Interface: OCI, a standard for packaging and executing containers
Dockerfile: A file describing a series of steps for building a container

* Container Orchestration: A way of managing and running  a set of containers

* Kubernetes:  The main container orchestration runtime

* Docker compose: A way of running a group of containers described by yaml (comes with docker)	

* Registry: A server for downloading containers for use by a container runtime (docker hub)!

### Model Serving Terms

* Prediction: Output from a machine learning model

* Model: A mathematical algorithm representing a static state mapping inputs to outputs.

* Model file: A description of a model to be loaded by code for execution

* Input data: A set of data points (images, csv,..) to be predicted on by a model

* Inference: The actual action of applying input data to a model and getting an output

* Text data: Data in plain text (human readable)

* Binary data: Data that is not human readable (zip files, images,..)

* Pre processing/Vectorization: A series of steps for transforming raw data in to a matrix usable by a model to get output from

* Pipeline: A combination of a pre processing step and a model to get output from a model

* Model Server: A server that runs 1 or more models as a server allowing remote access to model output from a client

* Transactional Serving: A way of running model output such that each input waits for a corresponding output

* Batch Serving: Runs a large set of inputs on a model (generally in a distributed system) in offline mode (saves the results to a specified output location rather than waiting for a response)!
